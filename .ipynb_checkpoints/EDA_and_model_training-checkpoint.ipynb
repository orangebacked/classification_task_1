## 1. Data exploration and correlation analysis

```python
# initial imports
import pandas as pd
import sklearn
import numpy as np
from scipy.stats import chi2_contingency 
import numpy as np


%matplotlib inline
def calculate_chi2_contigency(df:pd.DataFrame, namevar1:str, namevar2:str) -> str:
    chisqt = pd.crosstab(df[namevar1], df[namevar2], margins=True)
    chisqt = pd.crosstab(df[namevar1], df[namevar2], margins=True)
    value = np.array([chisqt.iloc[0][0:5].values,
                      chisqt.iloc[1][0:5].values])
    value_test = chi2_contingency(value)[0:3][0]
    p_val = chi2_contingency(value)[0:3][1]
    degrees_freedom = chi2_contingency(value)[0:3][2]
    
    return {"value":  value_test, "P-val": p_val, "degrees_freedom": degrees_freedom}

def evaluate_models(metric_ = "f1", list_algos = [LinearSVC, GaussianNB, KNeighborsClassifier, RandomForestClassifier, AdaBoostClassifier], cv_ = 10):
    list_dicts = []
    for algorithms in list_algos:
        dict_results = {}
        SVC_estimator = make_pipeline(preprocessing, algorithms())
        SVC_estimator.fit(train_set, y)
        y_scores = cross_val_score(SVC_estimator,  train_set, y, cv=cv_, scoring=metric_)
        dict_results["algorithm_name"] = algorithms.__name__
        dict_results.update(pd.Series(y_scores).describe())
        dict_results["metric"] = metric_
        list_dicts.append(dict_results)
    
    return pd.DataFrame(list_dicts)

```

__How are the continuous variables distributed?___

sold_quantity, initial quantity, price and days_passed are distributed with long tail distributions. Because sold_quantity and initial_quantity are so concentrated in a single value it was best to make them discrete thus the creation of the bin variables.

Sold quantities bigger that 0 are more correlated to new items and initial quantities bigger that 1 are also skewed towards new items.Because the distribution of initial quantities and sold quantities are so concentrated in a value we are going to binerize it into 0 and ones given this insight.

Moreover, the variables are not very correlated between each other which is good because it gives us exogenous variations for the model.

```python
#chi2 test
list_vals = [calculate_chi2_contigency(train_set, "condition",var) for var in ["shipping","accepts_mercadopago", "status", "listing_type_id"]]
pd.DataFrame.from_dict(list_vals)
```

```python
# binirized initial_quantity and sold_quantity
for var in ["bin_initial_quantity","bin_sold_quantity"]:
    tup = calculate_chi2_contigency(train_set, "condition",var)
    print(var, "and condition chi2test",tup )
    print("\n")
    
```

All the categorical values are correlated with the target variable, and, applying a chi2 test, we can reject the null hypothesis that the target variable and the feature are completely independent. Additionally, using the binirized initial quantity and sold quantity we can see that they are highly correlated and can reject the null hypothesis that they are independent from the target variable.
